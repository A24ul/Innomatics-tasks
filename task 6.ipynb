1.Array

The NumPy (Numeric Python) package helps us manipulate large arrays and matrices of numeric data.

* To use the NumPy module, we need to import it using:
import numpy

* Arrays
A NumPy array is a grid of values. They are similar to lists, except that every element of an array must be the same type.

```python
import numpy
def arrays(arr):
    return(numpy.array(arr[::-1], float))
arr = input('enter numbers :').strip().split(' ')
result = arrays(arr)
print(result)
```

2.Shape and Reshape

```python
import numpy
A = numpy.array(input('Enter space-separated numbers : ').split(),int)
A = A.reshape(3,3)
print(A)
```

3.Transpose and Flatten

```python
import numpy
N,M = map(int, input('Enter space-separeted values of N and M : ').split())
A = numpy.array([input('Enter space-separated elements : ').split() for _ in range(N)],int) # Enter elements for each row. 
print(A.transpose())
print(A.flatten())
```

5.Zeros and Ones

7.Array Mathematics

```python
import numpy
N,M = map(int, input('Enter space-separeted values of M and N : ').split()) # N lines and M integer/line
A = numpy.array([list(map(int, input('Enter space-separated elements : ').split())) for n in range(N)])
B = numpy.array([list(map(int, input('Enter space-separated elements : ').split())) for n in range(N)])

print(A+B)
print(A-B)
print(A*B)
print(A//B)
print(A%B)
print(A**B)
```

8.Floor,Ceil and Rint

```python
import numpy
A = numpy.array([list(map(float, input('Enter space-separated elements : ').split()))])
print(numpy.floor(A))
print(numpy.ceil(A))
print(numpy.rint(A))
```

9.Sum and Prod

```python
import numpy
N, M = map(int, input('Enter space-separated values of N and M : ').split()) # N lines and M integers/line
A = numpy.array([input('Enter your numbers : ').split() for _ in range(N)], int)
Sum = numpy.sum(A,axis = 0)
Prod = numpy.prod(Sum,axis = 0)
print(Prod)
```

10.Min and Max

```python
import numpy
N, M = map(int, input('Enter space-separated values of N and M : ').split())
A = numpy.array([input('Enter your numbers : ').split() for _ in range(N)], int)
Min = numpy.min(A,axis = 1)
print(numpy.max(Min))
```

11.Mean, Var and Std

```python
import numpy
N, M = map(int, input('Enter space-separated values of N and M : ').split())
A = numpy.array([input('Enter your numbers : ').split() for _ in range(N)], int)
print(numpy.mean(A,axis = 1))
print(numpy.var(A,axis = 0))
numpy.set_printoptions(legacy='1.13')
print(numpy.std(A))
```

12.Dot and Cross

13.Inner and Outer

14.Polynomial
