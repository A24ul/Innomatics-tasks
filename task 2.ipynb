1. List Comprehensions

Q : You are given three integers x, y and z representing the dimensions of a cuboid along with an integer . Print a list of all possible coordinates given by(i,j,k) on a 3D grid where the sum of i+j+k  is not equal to n. 

```python
x = int(input('Enter x value :'))
y = int(input('Enter Y value :'))
z = int(input('Enter Z value :'))
n = int(input('Enter Number :'))
cp=[[i,j,k] for i in range(x+1)  for j in range (y+1) for k in range (z+1) if (i+j+k)!=n]
print(cp)
```

3. Nested Lists

Q: Given the names and grades for each student in a class of N  students, store them in a nested list and print the name(s) of any student(s) having the second lowest grade.

* Note: If there are multiple students with the second lowest grade, order their names alphabetically and print each name on a new line.

```python
List = [[input('Enter name of the student : '),float(input('Enter marks of the student : '))] for i in range(int(input('Enter number of student : ')))]

s = sorted(set([i[1] for i in List]))
for name in sorted(i[0] for i in List if i[1] == s[1]):
    print('Student scoring second lowest grade is : ',name)
```

4. Finding the percentage

Q: The provided code stub will read in a dictionary containing key/value pairs of name:[marks] for a list of students. Print the average of the marks array for the student name provided, showing 2 places after the decimal.

```python
n = int(input('Enter No.of averages :'))
student_marks = {}
for _ in range(n):
    name, *line = input('Enter Name and  Numbers :').split()
    scores = list(map(float, line))
    student_marks[name] = scores
query_name = input('Enter Whose average you wish to know :')
total_marks = sum(student_marks[query_name])
average_marks = total_marks/3 
print("%.2f"%(average_marks))
```

5. Lists

Q : Consider a list (list = []). You can perform the following commands:

* insert i e: Insert integer  at position .
* print: Print the list.
* remove e: Delete the first occurrence of integer .
* append e: Insert integer  at the end of the list.
* sort: Sort the list.
* pop: Pop the last element from the list.
* reverse: Reverse the list.

Initialize your list and read in the value of N followed by  lines of commands where each command will be of the  types listed above. Iterate through each command in order and perform the corresponding operation on your list.

```python
N = int(input('Enter No.of operations You wiah to perform :'))
Output = [];
for i in range(0,N):
    lst = input().split();
    if lst[0] == "print":
        print(Output)
    elif lst[0] == "insert":
        Output.insert(int(lst[1]),int(lst[2]))
    elif lst[0] == "remove":
        Output.remove(int(lst[1]))
    elif lst[0] == "pop":
        Output.pop();
    elif lst[0] == "append":
        Output.append(int(lst[1]))
    elif lst[0] == "sort":
        Output.sort();
    else:
        Output.reverse();
```

11. Set .discard(), .remove() & .pop()

```python
n = int(input('Enter no of elements :'))
s = set(map(int, input('Elements are :').split()))
num = int(input('no of operations you would like to perform :'))
for i in range(num):
    A = input().split()
    if A[0]=="remove":
        s.remove(int(A[1]))
    elif A[0]=="discard":
        s.discard(int(A[1]))
    else :
        s.pop()
print('Elements remaining in set are : ', sum(list(s)))
```

16. Set Mutations

```python
len_set = int(input('Enter the length of Set :'))

S = set(map(int, input('Enter Elements :').split()))
N = int(input('Enter Number of other sets :'))

for i in range(N):
    print('Enter operation name and size of the set :')
    operation = input().split()
    if operation[0] == 'intersection_update':
        temp_storage = set(map(int, input('Enter Elements :').split()))
        S.intersection_update(temp_storage)
    elif operation[0] == 'update':
        temp_storage = set(map(int, input('Enter Elements :').split()))
        S.update(temp_storage)
    elif operation[0] == 'symmetric_difference_update':
        temp_storage = set(map(int, input('Enter Elements :').split()))
        S.symmetric_difference_update(temp_storage)
    elif operation[0] == 'difference_update':
        temp_storage = set(map(int, input('Enter Elements :').split()))
        S.difference_update(temp_storage)
    else :
        assert False

print('The sum of elements in set after these operations is ' ,sum(S))
```

19. Check Strict Superset 

```python
Main_Set = set(input('Enter number of elements separated by space : ').split())
N = int(input('Number of other set : '))
output = True
for i in range(N):
    Other_Set = set(input('Enter elements in other set : ').split())
    if not Other_Set.issubset(Main_Set):
        output = False
    if len(Other_Set) >= len(Main_Set):
        output = False

print('Is It Strict Superset : ' ,output)
```
