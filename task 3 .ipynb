1. Polar Coordinates 

```python
import cmath
Number = complex(input('Enter a complex Number : '))
Z = complex(Number)
print(cmath.polar(Z)[0])
print(cmath.polar(Z)[1])
```

1. Find Angle MBC

```python
import math

AB = float(input('Enter length of side AB : '))
BC = float(input('Enter length of side BC : '))
AC = math.sqrt((AB**2) + (BC**2))
H = AC/2
Adj = BC/2
Angle = int(round(math.degrees(math.acos(Adj/H))))
Angle = str(Angle)
sym = chr(176)
print('Angle MBC is :' ,Angle + sym )
```

1. Mod Divmod 

```python
A = int(input('Enter first number : '))
B = int(input('Enter Second number : '))
print('Integar division(Quotient) : ',A//B)
print('Modulus division(Remainder) : ',A%B)
print('Using divmod : ',divmod(A,B))
```

1. Power - Mod Power

```python
A = int(input('Enter first number : '))
B = int(input('Enter Second number : '))
C = int(input('Enter Third number : ')) 
print(pow(A,B)) 
print(pow(A,B,C))
```

1. Intergers Comes in All Sizes

1. Triangle Quest
