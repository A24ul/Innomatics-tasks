4. Mutations

```python
def mutate_string(string, position, character):
    L = list(string)
    L[position] = character
    string = ''.join(L)
    return string

if __name__ == '__main__':
    s = input('Enter a string : ')
    i, c = input('Enter position and character with space : ').split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```

5. Find a String

```python
def count_substring(string, sub_string):
    count = 0
    for i in range(len(string) - len(sub_string) + 1):
        if (string[i : i + len(sub_string)] == sub_string):
            count = count + 1
    return count

if __name__ == '__main__':
    string = input('Enter your main string : ').strip()
    sub_string = input('Enter your sub-string : ').strip()
    count = count_substring(string, sub_string)
    print('Number of repetition :',count)
```

6. String Validators

```python
def fun1(s):
    for i in range(len(s)):
        if(s[i].isalnum()):
            return True;
            break;
    return False;
        
def fun2(s):
    for i in range(len(s)):
        if(s[i].isalpha()):
            return True;
            break;
    return False;

def fun3(s):
    for i in range(len(s)):
        if(s[i].isdigit()):
            return True;
            break;
    return False;

def fun4(s):
    for i in range(len(s)):
        if(s[i].islower()):
            return True;
            break;
    return False; 
     
def fun5(s):
    for i in range(len(s)):
        if(s[i].isupper()):
            return True;
            break;
    return False;
    
if __name__ == '__main__':
    s = input('Enter String :')
    flagalphanum = fun1(s)
    alphabetical = fun2(s)
    digits = fun3(s)
    lowercase = fun4(s)
    uppercase = fun5(s)
    print(flagalphanum)
    print(alphabetical)
    print(digits)
    print(lowercase)
    print(uppercase)
```

7. Text Alignment

```python
thickness = int(input('Enter width : ')) #This must be an odd number
c = input('Enter Charcter :')

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

8. Text Wrap 

```python
import textwrap

def wrap(string, max_width):
    a = textwrap.fill(string, max_width)
    return a

if __name__ == '__main__':
    string  = input('Enter your string : ')
    max_width = int(input('Enter width : '))
    result = wrap(string, max_width)
    print(result)
```

10. String Formatting

```python
def print_formatted(number):
    Length = len(bin(number)[2:]) 
    for i in range(1 , number + 1): 
        print(str(i).rjust(Length , ' ') , end = " ")
        print(oct(i)[2:].rjust(Length,' ') , end = " ")
        print(((hex(i)[2:]).upper()).rjust(Length , ' ') , end = " ")
        print(bin(i)[2:].rjust(Length , ' ') , end = " ")
        print()
        
if __name__ == '__main__':
    n = int(input('Enter your number : '))
    print_formatted(n)
```

11. Alphabet Rangoli

```python
def print_rangoli(size):
    Breadth  = size * 4 - 3
    String = ''
    for i in range(1 , size + 1):
        for j in range(0 , i):
            String = String + chr(96 + size - j)
            if len(String) < Breadth:
                String = String + '-'
        for k in range(i - 1 , 0 , -1):    
            String = String + chr(97 + size - k)
            if len(String) < Breadth:
                String = String + '-'
        print(String.center(Breadth , '-'))
        String = ''
    for i in range(size - 1 , 0 , -1):
        String = ''
        for j in range(0,i):
            String = String + chr(96 + size - j)
            if len(String) < Breadth:
                String = String + '-'
        for k in range(i - 1 , 0 , -1):
            String = String + chr(97 + size - k)
            if len(String) < Breadth:
                String = String + '-'
        print(String.center(Breadth,'-'))
        
if __name__ == '__main__':
    N = int(input('Enter your number : '))
    print_rangoli(N)
```

14. Merge the Tools!

```python
from collections import OrderedDict

def merge_the_tools(string, k):
    Length = len(string)
    for i in range(0 , Length , k):
        print(' '.join(OrderedDict.fromkeys(string[i : i + k])))

if __name__ == '__main__':
    string = input('Enter your string : ')
    k= int(input('Enter integer : '))
    merge_the_tools(string, k)
```
