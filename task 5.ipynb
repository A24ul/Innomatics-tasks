4. Re.findall() & Re.finditer()

* re.findall():
The expression re.findall() returns all the non-overlapping matches of patterns in a string as a list of strings.
* re.finditer():
The expression re.finditer() returns an iterator yielding MatchObject instances over all non-overlapping matches for the re pattern in the string.

```python
import re
String = re.findall(r'(?<=[qwrtypsdfghjklzxcvbnm])([aeiou]{2,})(?=[qwrtypsdfghjklzxcvbnm])', input('Enter your string : ').strip(), re.IGNORECASE) # For case-sensitive

if String:
    for i in String:
        print(i)
else:
    print(-1)
```

5. Regex Substitution

* For every 'N' lines, you have to modify something into it.

```python
import re

def Word(any_match):
    if any_match.group(1) == '&&':
        return 'and'
    else:
        return 'or'

for _ in range(int(input('Enter number of lines : '))):
    print(re.sub(r"(?<= )(\|\||&&)(?= )", Word,input('Enter your string : ')))
```

11. HTML Parser - Part 1
* HTML :
 Hypertext Markup Language is a standard markup language used for creating World Wide Web pages.

* Parsing :
Parsing is the process of syntactic analysis of a string of symbols. It involves resolving a string into its component parts and describing their syntactic roles.

* HTMLParser :
An HTMLParser instance is fed HTML data and calls handler methods when start tags, end tags, text, comments, and other markup elements are encountered.

```python
from html.parser import HTMLParser
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print ('Start :', tag)
        for ele in attrs:
            print ('->', ele[0], '>', ele[1])

    def handle_endtag(self, tag):
        print ('End   :', tag)

    def handle_startendtag(self, tag, attrs):
        print ('Empty :', tag)
        for ele in attrs:
            print ('->', ele[0], '>', ele[1])

HTMLParser = MyHTMLParser()
for _ in range(int(input('Enter number of lines : '))):
    HTMLParser.feed(input('Enter your HTML Code : '))
```

12. HTML Parser - Part 2

* You are given an HTML code snippet of N lines.
Your task is to print the single-line comments, multi-line comments and the data.

```python
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_comment(self, data):
        if (len(data.split('\n')) != 1):
            print(">>> Multi-line Comment")
        else:
            print(">>> Single-line Comment")
        print(data.replace("\r", "\n"))
    def handle_data(self, data):
        if data.strip():
            print(">>> Data")
            print(data)
Html = " "       
for i in range(int(input('Enter number of lines : '))):
    Html = Html + input('Enter your HTML Code : ').rstrip()
    Html = Html + '\n'
    
parser = MyHTMLParser()
parser.feed(Html)
parser.close()
```

```python
from html.parser import HTMLParser
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print(tag)
        for attr, value in attrs:
            print("->", attr, ">", value)

    def handle_startendtag(self, tag, attrs):
        print(tag)
        for attr, value in attrs:
            print("->", attr, ">", value)

Html = ''
for _ in range(int(input('Enter number of lines : '))):
    Html = Html + input('Enter your HTML Code : ').rstrip() + '\n'

parser = MyHTMLParser()
parser.feed(Html)
parser.close()
```

14. Validating UID

* We have to validate all randomly generated Unique Identification Number (UID)

```python
import re

for i in range(int(input('Enter number of lines : '))):
    N = input('Enter your UID : ').strip()
    if N.isalnum() and len(N) == 10:
        if bool(re.search(r'(.*[A-Z]){2,}',N)) and bool(re.search(r'(.*[0-9]){3,}',N)):
            if re.search(r'.*(.).*\1+.*',N):
                print('Invalid')
            else:
                print('Valid')
        else:
            print('Invalid')
    else:
        print('Invalid')
```

15. Validating Credit Card Numbers 

You and Fredrick are good friends. Yesterday, Fredrick received N credit cards from ABCD Bank. He wants to verify whether his credit card numbers are valid or not.

```python
import re
for i in range(int(input('Enter number of lines : '))):
    S = input('Enter your credit card details : ').strip()
    pre_match = re.search(r'^[456]\d{3}(-?)\d{4}\1\d{4}\1\d{4}$',S)
    if pre_match:
        processed_string = "".join(pre_match.group(0).split('-'))
        final_match = re.search(r'(\d)\1{3,}',processed_string)
        if final_match:
            print('Invalid')
        else :
            print('Valid')
    else:
        print('Invalid')
```

17. Matrix Script

```python
import math
import os
import random
import re
import sys

first_multiple_input = input('Enter number of rows and columns : ').rstrip().split()

N = int(first_multiple_input[0]) # Rows
M = int(first_multiple_input[1]) # Columns

Matrix = []

for _ in range(N):
    Matrix_item = input('Enter elements of your matrix : ')
    Matrix.append(Matrix_item)

Matrix = list(zip(*Matrix))
sample = str()

for words in Matrix:
    for char in words:
        sample += char
       
print(re.sub(r'(?<=\w)([^\w\d]+)(?=\w)', ' ', sample))
```
